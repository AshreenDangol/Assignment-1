***OUTLIERS DETECTION***

```python
#identifying the countries in south asia that are outliers based out of score and gdp per capita


#Calculating iqr for gdp

gdpQ1 = south_df['Log GDP per capita'].quantile(0.25)
gdpQ3 = south_df['Log GDP per capita'].quantile(0.75)
gdpIQR = gdpQ3 - gdpQ1


#for very low gdp
GdpLow = gdpQ1 - 1.5 * gdpIQR
#for high gdp

GdpHigh = gdpQ3 + 1.5 * gdpIQR

#for iqr score
scoreQ1 = south_df['score'].quantile(0.25)
scoreQ3= south_df['score'].quantile(0.75)
scoreIQR = scoreQ3 - scoreQ1
scoreLowGdp =scoreQ1 - 1.5 * scoreIQR
scoreHighGdp = scoreQ3 + 1.5 * scoreIQR

#identifying the outliers
#using | to combine multiple conditionss..
south_df['Outliers'] = (south_df['Log GDP per capita'] < GdpLow )|(south_df['Log GDP per capita'] < GdpHigh)| (south_df['score'] < scoreLowGdp)| (south_df['score'] < scoreHighGdp)
plt.figure(figsize=(10, 6))
#creating scatterplot for visualization
plt.scatter(south_df['Log GDP per capita'], south_df['score'],
            c=south_df['Outliers'].map({True: 'red', False:'blue'}),
            label ='data points'

)
plt.xlabel('Log GDP per capita')
plt.ylabel('score')
plt.title('Outliers Detection')
plt.legend()
```

it shows the country with extemrely high and low gdp per capita..
the countries that have significant deviation in score has unique social economic.for high gdp- highh outliers it  inflates the regional gdp and shows its area is more successful and for low outliers it deflates the gdp and its all time low

**Outlier dewtection**

```python
#looping through each region and its corresponding datas
for region, mixed_df in [('South Asia', south_df), ('Middle East', middle_df)]:
    #getting the 25th percentile and the 75 percentile
    ScoreQ1 = mixed_df['score'].quantile(0.25)
    ScoreQ3 = mixed_df['score'].quantile(0.75)
    #Interquartile range for Score
    ScoreIQR = ScoreQ3 - ScoreQ1

    gdpQ1 = mixed_df['Log GDP per capita'].quantile(0.25)
    gdpQ3 = mixed_df['Log GDP per capita'].quantile(0.75)

    #Interquartile range for GDP
    gdpIQR = gdpQ3 - gdpQ1

    #Calculating the bounds to detect outliers
    scoreLowerBound = ScoreQ1 - 1.5 * ScoreIQR
    scoreUpperBound = ScoreQ3 + 1.5 * ScoreIQR
    gdpLowerBound = gdpQ1 - 1.5 * gdpIQR
    gdpUpperBound = gdpQ3 + 1.5 * gdpIQR

    #it shows the rows where Score or GDP are outside the acceptable range
    mixed_df['Outliers'] = (
        (mixed_df['score'] < scoreLowerBound) | (mixed_df['score'] > scoreUpperBound) |
        (mixed_df['Log GDP per capita'] < gdpLowerBound) | (mixed_df['Log GDP per capita'] > gdpUpperBound)
    )

    #creating the scatter plot for visualizing
    plt.figure(figsize=(10, 6))
    plt.scatter(mixed_df['Log GDP per capita'], mixed_df['score'],
                c=mixed_df['Outliers'].map({True: 'red', False: 'blue'}))
    plt.xlabel('Log GDP per capita')
    plt.ylabel('Score')
    plt.title(f'Outliers Detection for {region}')
    plt.legend(['Outliers (red)', 'Normal data (blue)'])
    plt.show()


```

in case 1 high score low gdp,
  the countries that fall under this might indicate a high happiness or satisfaction even if the income is low, which suggets the culture, social factors that positively influences the happiness

in case 2 low score, high gdp,
  the countries that fall under this might indicate a high income so high economy but low happiness due to economy unequality, social unrest..
in case 3 extreme values
  in this case countires with either extremely low or high gdp and extremely low or high score might show unique geopolitical and social scenarios
